```python
import requests
from bs4 import BeautifulSoup

def search_news_urls(keyword, num_results):
    search_engine = "https://www.google.com/search"

    # Create a dictionary of query parameters
    params = {
        "q": keyword + " news",
        "num": num_results,
        "hl": "en",  # Language can be adjusted as needed
        "tbm": "nws",  # Restrict search results to news articles
    }

    # Send an HTTP GET request to the search engine
    response = requests.get(search_engine, params=params)

    # Parse the HTML content of the search results page
    soup = BeautifulSoup(response.text, "html.parser")

    # Find and extract the URLs from the search results
    urls = []
    for link in soup.find_all("a", href=True):
        url = link["href"]
        if url.startswith("/url?q="):
            # Extract the actual URL from the Google search result format
            url = url.split("/url?q=")[1].split("&")[0]
            urls.append(url)

    return urls

# Main function
if __name__ == "__main__":
    keyword = input("Enter the keyword to search for news articles: ")
    num_results = int(input("Enter the number of results to retrieve: "))

    news_urls = search_news_urls(keyword, num_results)
    lst=[]
    print(f"News articles related to '{keyword}':")
    for i, url in enumerate(news_urls, start=1 ):
        print(f"{i}. {url}")
        lst.append(url)
        if(i==num_results):
          break
```

Cosine Similarity


```python
import math
import string
import sys
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
def read_file(filename):

	try:
		with open(filename, 'r') as f:
			data = f.read()
		return data

	except IOError:
		print("Error opening or reading input file: ", filename)
		sys.exit()
```

```python
translation_table = str.maketrans(string.punctuation+string.ascii_uppercase,
									" "*len(string.punctuation)+string.ascii_lowercase)

def get_words(text):

	text = text.translate(translation_table)
	word_list = text.split()

	return word_list

def count_frequency(word_list):

	D = {}

	for new_word in word_list:

		if new_word in D:
			D[new_word] = D[new_word] + 1

		else:
			D[new_word] = 1

	return D
```

```python
def word_frequencies(filename):

	line_list = read_file(filename)
	word_list = get_words(line_list)
	freq_mapping = count_frequency(word_list)

	print("File", filename, ":", )
	return freq_mapping

def dotProduct(D1, D2):
	Sum = 0.0

	for key in D1:

		if key in D2:
			Sum += (D1[key] * D2[key])

	return Sum

def vector_angle(D1, D2):
	numerator = dotProduct(D1, D2)
	denominator = math.sqrt(dotProduct(D1, D1)*dotProduct(D2, D2))

	return math.acos(numerator / denominator)

def documentSimilarity(filename_1, filename_2):

	sorted_word_list_1 = word_frequencies(filename_1)
	sorted_word_list_2 = word_frequencies(filename_2)
	distance = vector_angle(sorted_word_list_1, sorted_word_list_2)
	if(distance>1):
		distance=distance-0.5
	print("The distance between the documents is: % 0.3f (radians)"% distance)
	print("The percentage similarity between", filename_1, "and" ,filename_2, "is: % 0.2f" % ((1-distance)*100),"%")

for i in range(2,11):
		documentSimilarity('NewsFile{}.txt'.format(1), 'NewsFile{}.txt'.format(i))
		print(" ")

# Define a function to calculate cosine similarity between two documents
def calculate_cosine_similarity(doc1, doc2):
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform([doc1, doc2])
    return cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]
```

Jaccard Similarity


```python
def jaccard_similarity(filename_1, filename_2):
    intersection = len(filename_1.intersection(filename_2))
    union = len(filename_1) + len(filename_2) - intersection
    similarity = intersection / union
    print(" The percentage similarity is: % 0.2f" % ((similarity)*100),"%")

def read_and_tokenize(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        tokens = set(text.lower().split())
    return tokens

for i in range(2,11):
   # File paths for the two text documents
   file1_path = 'NewsFile{}.txt'.format(1)
   file2_path = 'NewsFile{}.txt'.format(i)
   print(" For File", file1_path, "and", file2_path )
   tokens1 = read_and_tokenize(file1_path)
   tokens2 = read_and_tokenize(file2_path)
   # Calculate Jaccard similarity
   similarity = jaccard_similarity(tokens1, tokens2)
   print("")
```

Using a combination of TF-IDF (Term Frequency-Inverse Document Frequency) and Word2Vec-based similarity

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
from gensim.models import Word2Vec
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load spaCy model
nlp = spacy.load('en_core_web_sm')

def preprocess_text(text):
    # Tokenize
    tokens = word_tokenize(text.lower())

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word.isalnum() and word not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return tokens

def get_similarity(doc1, doc2):
    # Tokenize, remove stopwords, and lemmatize
    tokens1 = preprocess_text(doc1)
    tokens2 = preprocess_text(doc2)

    # Compute TF-IDF vectors
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform([" ".join(tokens1), " ".join(tokens2)])

    # Compute cosine similarity based on TF-IDF vectors
    tfidf_similarity = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]

    # Train Word2Vec model
    model = Word2Vec([tokens1, tokens2], vector_size=100, window=5, min_count=1, workers=4)

    # Filter out words that are not in the model's vocabulary
    tokens1_filtered = [word for word in tokens1 if word in model.wv.key_to_index]
    tokens2_filtered = [word for word in tokens2 if word in model.wv.key_to_index]

    # Check if there are words left after filtering
    if not tokens1_filtered or not tokens2_filtered:
        return 0.0

    # Compute Word2Vec similarity
    w2v_similarity = model.wv.n_similarity(tokens1_filtered, tokens2_filtered)

    # Combine similarities
    combined_similarity = 0.5 * tfidf_similarity + 0.5 * w2v_similarity

    return combined_similarity



def read_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    return content

word2vec_value=[]
for i in range(2,11):
  # File paths for the two text documents
  file1_path = 'NewsFile{}.txt'.format(1)
  file2_path = 'NewsFile{}.txt'.format(i)
  print("For File", file1_path, "and", file2_path)
  doc1 = read_file(file1_path)
  doc2 = read_file(file2_path)
  similarity_score = get_similarity(doc1, doc2)
  word2vec_value.append(similarity_score)
  print("Combined Similarity Score% 0.4f:" %(similarity_score))
```

Combination of BERT and Cosine Similarity


```python
import tensorflow as tf
from transformers import BertTokenizer, TFBertModel
from sklearn.metrics.pairwise import cosine_similarity

# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
bert_model = TFBertModel.from_pretrained("bert-base-uncased")

def get_bert_embeddings(text):
    # Tokenize input text
    tokens = tokenizer(text, return_tensors="tf", truncation=True, padding=True)

    # Get BERT model output (embeddings)
    output = bert_model(**tokens)
    embeddings = output.last_hidden_state[:, 0, :].numpy()

    return embeddings

def cosine_similarity_bert(doc1, doc2):
    # Get BERT embeddings for both documents
    embeddings1 = get_bert_embeddings(doc1)
    embeddings2 = get_bert_embeddings(doc2)

    # Compute cosine similarity between embeddings
    similarity_score = cosine_similarity(embeddings1, embeddings2)[0][0]

    return similarity_score

def read_and_tokenize(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        # Tokenize the text into a list of strings
        tokens = text.lower().split()
    return tokens

bert_value=[]

for i in range(2,11):
        # File paths for the two text documents
        file1_path = 'NewsFile{}.txt'.format(1)
        file2_path = 'NewsFile{}.txt'.format(i)
        print("For File", file1_path, "and", file2_path)

        # Read and tokenize the text from the files
        tokens1 = read_and_tokenize(file1_path)
        tokens2 = read_and_tokenize(file2_path)

        # Join the tokens into a single string
        text1 = ' '.join(tokens1)
        text2 = ' '.join(tokens2)

        # Calculate the combined similarity score

        similarity_score = cosine_similarity_bert(text1, text2)
        bert_value.append(similarity_score)
        print("Combined Similarity Score: %0.4f" %(similarity_score))
```
